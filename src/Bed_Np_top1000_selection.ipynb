Greetings! For the test assignment I chose YY1 TF data from Homo sapiens sample GM12892. See link https://www.encodeproject.org/experiments/ENCSR000BLT/ for more details.

In this notebook you will find how to select top 1000 peaks based on qValue. Why choose qValue? 1.ENCODE preocessed and released data have pValue column unassigned and filled with -1, so we are unable to choose top peaks. In the test assignment pValue was the quality metrics for selection, but column values are useless(thanks to ENCODE and/or researchers who released data this way). 2.On the other hand, qValue column filled properly. It is measurement of statistical significance using false discovery rate (-log10). To put in simple, the more is qValue in result file (narrowPeak .bed), the more significant peaks are. See http://genome.ucsc.edu/FAQ/FAQformat.html#format12 for more details.
Citing from here:https://www.biorxiv.org/content/biorxiv/suppl/2017/04/10/120113.DC2/120113-1.pdf
A peak call with a larger signal value has a smaller p- and q-value, which indicates that it is more likely to reflect an actual protein-DNA binding event. Thus, a larger signal value will translate to a larger − log10(p-value) and − log10(q-value).
